```python
import pandas as pd
url='https://raw.githubusercontent.com/coding-blocks-archives/machine-learning-online-2018/master/13.%20Decision%20Trees/titanic.csv'
data = pd.read_csv(url)
clean=['Name','PassengerId','Parch','Cabin','Embarked','Ticket']
data_clean=data.drop(clean,axis=1)
data_clean.info()
```

```python
dic={'male':0,'female':1}
data_clean.Sex=data_clean.Sex.map(dic)
data_clean.Sex
#data_clean=data_clean.fillna(data_clean.Age.mean())
data_clean.describe()


X=data_clean[['Sex','Fare']]
y=data_clean['Survived']
```

```python
X
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[1,7]])
```

```python
from matplotlib import pyplot as plt
score=[KNeighborsClassifier(n_neighbors=i).fit(X,y).score(X,y) for i in range(3,16)]

plt.plot(range(3,16),score)
plt.show()
```
