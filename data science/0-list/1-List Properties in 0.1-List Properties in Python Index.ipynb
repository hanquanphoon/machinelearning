## Find the Properties of List
<br>
<div style="font-size:16px; line-height:25pt">
How can we find the properties of a list? To find the properties, we will use the methods below:
<br>
<div style="line-height:20pt">
To find the index of a number or find a number through index
<ul>
	<li><code>data[index]</code></li>
	<li><code>data.index()</code></li>
    <li><code>numpy.where()</code> for numpy arrays</li>
</ul>
</div>
Have a look at the example below to get a better understanding.
</div>

```python
x = [7, 14, 20, 1, -16, -10, -8, -3, -12, 2, -1, -19, 2, 1, 9]

# To find a number by index
print 'The number with index 10: ', (x[10])
#   data_name[index_number]


# To find the index of an items
print 'The index of number 1 in the range of 0 to 5 is: ', (x.index(1,0,5))
#     data_name.index(item to be found, start, end)
# to start and end in a specific range of index

import numpy as np
x_arrays = np.array([7, 14, 20, 1, -16, -10, -8, -3, -12, 2, -1, -19, 2, 1, 9])
print 'The index of number of -12: ', (np.where(x_arrays == -12))
#                                      np.where(list_name == item to be found)
```

<div style="font-size:16px; line-height:20pt">
To find the length, max value, min value and sum of items for a list.
<ul>
    <li><code>len()</code></li>
    <li><code>numpy.size</code> for numpy arrays</li>
    <li><code>max()</code></li>
    <li><code>min()</code></li>
    <li><code>sum()</code></li>
</ul>
<br>
Have a look at the example below to get a better understanding.
</div>

```python
x = [7, 14, 20, 1, -16, -10, -8, -3, -12, 2, -1, -19, 2, 1, 9]

# To find the length of list
print 'length of list: ', len(x)
#     len(data_name)


# To find the maximum value of list
print 'maximum value of list: ', max(x)
#     max(data_name)


# To find the minimum value of list
print 'minimum value of list: ', min(x)
#     min(data_name)


# To find the sum of all items in the list
print 'sum of all items in the list: ', sum(x)
#     sum(data_name)


import numpy as np
x_array = np.array([7, 14, 20, 1, -16, -10, -8, -3, -12, 2, -1, -19, 2, 1, 9])
print 'Size / Length of array: ', x_array.size
```
