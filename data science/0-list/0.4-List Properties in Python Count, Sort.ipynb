# <u>Count function</u>
<div style="font-size:16px; line-height:25pt">
    <code>list.count()</code> function is use to counts how many times an element has occurred in a list and returns it.
</div>
<br><br>

## How to use Count function in Python and Numpy
<br>
<div style="font-size:16px; line-height:25pt">
    <ul style="list-style-type:none;font-size:25px;">
        <li><u><b>In Python</b></u></li>
        <ol style="font-size:16px;">
            <li><code>list.count(x)</code> - Return the number of times x appears in the list.</li>
        </ol><br>
        <li><u><b>In Numpy</b></u></li>
        <ol style="font-size:16px; list-style-type:lower-alpha;">
            <li><code>numpy.count_nonzero(array == x_value)</code> - Return the number of times x appears in the array.</li>
        </ol>
    </ul><br>
    Look at the example below:
    <h3>In Python</h3>
</div>

```python
data1 = []
print(data1.count(1),'times of 1 appear in the list')
```

<div style="font-size:16px; line-height:25pt">
    <h3>In Numpy</h3>
</div>

```python
import numpy as np
data2 = np.array([5, 1, 9, 4, 9, 2, 3, 5, 9, 7, 9, 3, 9, 3, 5, 2, 1, 5, 4, 2])
print(np.count_nonzero(data2 == 9),'times of 9 appear in the list')
```

# <u>Sort function</u>
<div style="font-size:16px; line-height:25pt">
    <code>list.sort()</code> function is use to sort a list in ascending, descending or user defined order.
</div>
<br><br>

## How to use Sort function in Python and Numpy
<br>
<div style="font-size:16px; line-height:25pt">
    <ul style="list-style-type:none;font-size:25px;">
        <li><u><b>In Python</b></u></li>
        <ol style="font-size:16px;">
            <li><code>list.sort(key=None, reverse=False)</code> - Sort the items of the list in place (the arguments can be used for sort customization, see sorted() for their explanation).</li>
        </ol><br>
        <li><u><b>In Numpy</b></u></li>
        <ol style="font-size:16px; list-style-type:lower-alpha;">
            <li><code>numpy.sort(a, axis=-1, kind=None, order=None)</code> - Return a sorted copy of an array.</li>
        </ol>
    </ul><br>
    Look at the example below:
    <h3>In Python</h3>
</div>

```python
data1 = [6, 3, 1, 4, 6, 5, 9, 2, 9, 4, 7, 10, 7, 3, 1, 3, 3, 7, 1, 1]
data1.sort()
print('Data1 is arranged in an accending oder', data1)

print('\n\n')

desc_data1 = data1.copy()
desc_data1.sort(reverse=True)
print('Desc_data1 is arranged in an descending oder', desc_data1)
```

<div style="font-size:16px; line-height:25pt">
    <h3>In Numpy</h3>
</div>

```python
import numpy as np
data2 = np.array([5, 1, 9, 4, 9, 2, 3, 5, 9, 7, 9, 3, 9, 3, 5, 2, 1, 5, 4, 2])
data2 = np.sort(data2)
print('Data1 is arranged in an accending oder', data2)
```
