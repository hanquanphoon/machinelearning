## <b>1.0 General explanation about <font color='deeppink'>Linear Regression</font> </b>

<font size =3 >Regression is the using of the relationship between variables to find best fit line or best equation. Best fit line is used to get the most accurate data which from there, we can make the best prediction. Linear regression is used for continuous data, or two correlated variables.

![alt text](https://cdn-images-1.medium.com/max/1200/1*xz_haBuJRSI2DaveNk-3gw.gif)
$$ Y = a_1X_1 + b (randomError) $$

## <b>1.1 Method used to calculate <font color='deeppink'>Linear Regression</font> line.</b> 

# $$ y = A + Bx $$

<font size =4>Method used: </font><font size =4 color='orangered'><b>LEAST SQUARE METHOD</b></font><br><br>

- <font size=4> A method to estimate the true value of some quantity <b>based on a consideration of errors </b>in observations or measurements.</font><br><br>

- <font size=4> It provides the <b>overall rationale placement of the line of the best fit</b> among the data points being studied. </font><br><br>

- <font size =4>AIM:<b> To calculate the values B (slope) and A (y-intercept) in the equation of a line.</b></font>


##  1.2 <font color='orangered'>Least Square Method </font>Explanation <br>
- <font size =4> It finds <b> all possible line equations </b>for a given set of data points and select the one <b>which has the least squared sum of errors (residual).</b></font><br><br>

- <font size =4><b>ERRORS :</b> the <b>difference</b> between the an observed value <font color='red'><b>(actual)</b></font> and the fitted value provided by a model <font color='red'><b>(predicted)</b></font></font><br><br>

- <font size =4> So our equation will produce <b>the value of y with the least value of errors </b></font> <br><br>

![alt_text](https://muthu.co/wp-content/uploads/2018/06/Snip20180616_19.png)


## <b>2.0 The manual steps to find the <font color="orangered">Linear Regression:</font>


<font size =4 >1. You have to build the <b>new equation </b>which consists of the least error values </font>

<font size =4 >2. So, find the <b>new slope (B)</b></font>

## $$ B = \frac {n\sum xy - \sum x \sum y} { n\sum x^2 - (\sum x)^2} $$

<font size =4 >3. Then, calculate the <b>y-intercept </b></font>

## $$ A = \frac {n\sum y - B\sum x} { n} $$

<font size =4>So, in order to calculate the values of B and A, you first need :</font>
## $x^2$, $y^2$, $xy$
## $\sum x$, $\sum y$, $\sum x^2$, $\sum y^2$, $\sum xy$<br>

## 3.0 Finding Linear Regression equation <font color="deeppink">using Python.</font>

## <font color='blue'>STEP 1:</font>

<font size =4> Store your data inside a table, which consists of column $x$ and $y$. Then, add 3 more columns inside the table labelled $xy$, $x^2$, and $y^2$. Note that $xy$ is the product of $x$ and $y$, $x^2$ is the square of $x$, and $y^2$ is the square of $y$.</font><br>

<font size =4> As we are using python to do the calculation, you can use list to store the values of x and y. It will be easier and faster to handle the data. </font>

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.arange(0,10,0.1)   # generate random value of x
y= (x*2) + (np.random.normal(0,1,len(x))*3)   # generate random value of x

plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x,y)

## Step 1
xy=[]
x2=[]
y2=[]   # create an empty list for each of the new column

for m in range(len(x)):
    ab = x[m] * y[m]   # calculate x*y
    xy.append(ab)      # append each value of x*y into the list
# print("xy: "+ str(xy))

for i in x:
    a= i**2        # calculate x power of 2
    x2.append(a)   # append each value of x^2 into the list
# print("x2: "+ str(x2))

for j in y:
    b= j**2        # calculate y power of 2
    y2.append(b)   # append each value of y^2 into the list
# print("y2: "+ str(y2))

```

## <font color='blue'>STEP 2:</font>
<font size =4> Now add a new row to the bottom of your table. This row should store the total sum of each of column $x$, $y$, $xy$, $x^2$, and $y^2$. Use for loop to get the sum of each column. </font>


```python
## Step 2

nX=0
nY=0
nXY=0
nX2=0
nY2=0   # initialize the sum of each column as 0

for i in x:
    nX += i   # calculate sum of x
print("Sum of x: "+str(nX))   


for j in y:
    nY += j   # calculate sum of y
print("Sum of y: "+str(nY))


for k in xy:
    nXY += k   # calculate sum of xy
print("Sum of xy: "+str(nXY))


for m in x2:
    nX2 += m   # calculate sum of x2
print("Sum of x2: "+str(nX2))


for n in y2:   
    nY2 += n   # calculate sum of y2
print("Sum of y2: "+str(nY2))
```

## <font color='blue'>STEP 3:</font>

<font size =4>Using the following formula, calculate the value of $B$ (slope). Note that $n$ is the sample size (number of item) of the table. Make sure you translate the formula correctly for the calculation. </font>

## $$ B = \frac {n\sum xy - \sum x \sum y} { n\sum x^2 - (\sum x)^2} $$

<font size =4>Then, calculate the value of $A$ (intercept) using the formula which consists of $B$ that we calculated before. </font> 

## $$ A = \frac {n\sum y - B\sum x} { n} $$


```python
## Step 3
n= len(x)

B= 0   # initialize B and A to 0
A= 0

B = ((n*nXY) - (nX*nY)) / (n*nX2 - (nX**2))
A = ((nY) - B*(nX)) / n  # Translate the formula into calculation for A and B

print("B (slope)    : "+str(B))
print("A (intercept): "+str(A))
```

## <font color='blue'>STEP 4:</font>
<font size =4>Insert the values into the equation: </font>
## $y = A + Bx$
<font size =4> You can use for loop to generate the new $y$, for each value of $x$ to form the linear regression line. Store the value of new $y$ inside a list as we want to use it later on. </font>

```python
## Step 4

Y= []
for i in range(n):
    y0=  A + (B*x[i]) 
#     y0= A + B*x[i]
    Y.append(y0)
# print(Y)
```

## <font color='blue'>STEP 5:</font>
## Use 'plot(x, y)' function from matplotlib to form the linear regression line.

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(x,y)
plt.plot(x,Y, c="red")
```

## <font color='orchid'>STEP 2:</font>

```python
x= np.arange(0,10,0.1)
# generate data of x from 1-10 with difference of 0.1 between each data

y= 1 + (x*2) + (np.random.normal(0,1,len(x))*3)
# generate y based on x value using y=A + Bx + random noise (to make more realistic data)
```

```python
plt.xlabel('X')
plt.ylabel('Y')   # label the axis

plt.scatter(x,y)
# plot the points of each (x,y)
```
