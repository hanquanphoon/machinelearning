## 2.0 Example for <font color = "orange">Covariance</font>
Study was conducted to investigate the relationship between <font color = "red"><strong>score in Data Science</strong></font> and <font color = "red"><strong>weight in kilograms</strong></font> among students.

```python
#creating table using pandas library by extracting .csv file
import pandas as pd

pd.read_csv('1st.csv', delimiter = ',')
```

```python
#creating table using pandas library by inserting data
import pandas as pd

score = [3, 5, 8, 10, 13, 15, 18, 20, 28, 32, 35, 39, 40, 40, 42, 43, 45, 46, 47, 48]
weight = [30, 32, 34, 35, 38, 39, 41, 43, 44, 46, 47, 50, 51, 53, 54, 58, 60, 65, 66, 70]

question_table = pd.DataFrame(score, columns = ['Score ( /50)']); question_table['Weight (kg)'] = weight
question_table
question_table.style.set_properties(**{'text-align': 'center'})
```

### 2.1 Calculating the <font color = "orange">Covariance</font> step by step

```python
# score = [3, 5, 8, 10, 13, 15, 18, 20, 28, 32, 35, 39, 40, 40, 42, 43, 45, 46, 47, 48]
weight = [30, 32, 34, 35, 38, 39, 41, 43, 44, 46, 47, 50, 51, 53, 54, 58, 60, 65, 66, 70]

#1st Step: Find the mean of x and the mean of y
meanScore = sum(score) / len(score)
meanWeight = sum(weight) / len(weight)

print "1. Mean of score in Data Science among students is ", meanScore
print "2. Mean of weight (kg) among students is ", meanWeight

#2nd Step: Find the difference of each value of x (assign as a) and y (assign as b) from their means.
a = []
for i in score:
    differentFromMeanScore = i - meanScore
    a.append(differentFromMeanScore)
    #print "List of a is ", a

b = []
for j in weight:
    differentFromMeanWeight = j - meanWeight
    b.append(differentFromMeanWeight)
    #print "List of b is ", b
    
#3rd Step: Multiply the results obtained
ab = []
for k in range (len(a)):
    ab.append(a[k]*b[k])
    #print "List of (a x b) is ", ab

#4th Step: Find the covariance
covariance = sum(ab) / float(len(ab) - 1)
print "3. The covariance is ", covariance
```

### 2.2 Calculating the <font color = "orange">Covariance</font> using NUMPY

```python
import numpy as np

print "1. The covariance of (score, score) is ", np.cov(score)
print "2. The covariance of (weight, weight) is ", np.cov(weight)
print "3. The covariance of (score, weight) is "
print (np.cov(score, weight))
print "4. The covariance of (weight, score) is "
print (np.cov(weight, score))
```

```python
#1. Create the list of Days Present (as x) and GPA (as y)
x = [180, 150, 170, 180, 177, 180, 140, 169, 168, 152, 150, 170, 109, 180, 108, 175, 144, 120, 133, 160, 160, 159, 165, 180, 149, 180, 160, 155, 156, 145, 122, 131, 156, 166, 170, 155, 158, 170, 160, 165, 159, 119, 168, 175, 161, 160, 160, 150, 140, 148]
y = [4.00, 2.50, 4.00, 3.90, 3.75, 3.80, 2.90, 3.10, 3.25, 3.40, 3.30, 3.90, 1.35, 4.00, 1.00, 3.85, 2.98, 2.75, 2.75, 3.60, 3.50, 3.50, 3.50, 3.85, 2.95, 3.95, 3.65, 3.55, 3.58, 2.98, 1.50, 1.75, 2.20, 3.00, 3.00, 3.00, 3.15, 3.90, 3.15, 3.85, 2.70, 1.00, 3.25, 3.90, 2.80, 3.50, 3.40, 2.30, 2.50, 2.35]

#2.1 Find mean of x (as meanX) and mean of y (as meanY)
meanX = sum(x) / len (x)
meanY = sum(y) / len (y)

#2.2 Find the difference of each value of x form mean of x (as a), and each value in y from mean of y (as b)
a = []

for i in x:
    differenceFromMeanX = i - meanX
    a.append(differenceFromMeanX)

b = []

for j in y:
    differenceFromMeanY = j - meanY
    b.append(differenceFromMeanY)

#2.3 Find the product of the a and b (a x b)
ab = []

for k in range(len(a)):
    ab.append(a[k]*b[k])

#2.4 Find the covariance and print it
covariance = sum (ab) / float(len(ab) - 1)
print covariance
```

-  Product of the results obtain
-  Covariance and print it.

```python
#4. Create the list of Height and the Weight
height = [173.0, 173.0, 162.0, 163.0, 152.0, 172.0, 170.0, 171.0, 163.9, 151.0, 149.9, 149.9, 164.0, 180.0, 167.0]
weight = [50, 73, 53, 85, 74, 62, 75, 68, 65, 78, 50, 62, 57, 85, 51]

#5.1 Find mean of x, and mean of y
meanHeight = sum(height) / len(height)
meanWeight = sum(weight) / len(weight)

#5.2 Find the difference for each value of x from mean of x (as a), and each value of y from mean of y (as b)
c = []

for i in height:
    c.append(i - meanHeight)

d = []

for j in weight:
    d.append(j - meanWeight)
    
#5.3 Find the product of the results obtained
heightWeight = []

for k in range(len(c)):
    heightWeight.append(c[k]*d[k])

#5.4 Find the covariance and print it
covariance = sum(heightWeight) / float(len(heightWeight) - 1)
print covariance
```
