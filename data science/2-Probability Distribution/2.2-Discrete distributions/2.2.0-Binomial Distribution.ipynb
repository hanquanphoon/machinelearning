## <font color = "slateblue">2.0 Probability of an Event that follows a</font> <font color = "darkblue">Binomial Distribution</font>

### <font color = "darkcyan">2.1 Definition</font>

1.  The word <b><font color = "darkblue">Binomial</font></b> literally means "two number". A <b>finite number of trials</b> with only two outcomes <b>'success'</b> or <b>'failure'</b> is known as a Bernoulli Trial.</br>
2.  If a Bernoulli Trial is repeated many times, the experiment is known as the <b><font color = "darkblue">Binomial</font></b> experiment.</br>
3.  <b><font color = "darkblue">Binomial Distribution</font></b> is a discrete probability distribution. This means that the possible outcomes are distinct and non-overlapping.</br>
4.  Let X be a discrete random variable that represents the number of successes of binomial experiment. X follows a binomial distribution (with the number of experiments = <em>n</em>, the probability of success = <em>p</em> and is denoted by <b><em>X~B(n, p)</em></b></br>

### <font color = "darkcyan">2.2 Example 1</font>

<p>Now let's consider the binomial distribution for <b>100 trials</b>, or 100 coin flips. What if we actually wanted to calculate the probability of flipping a coin 100 times, and <b>getting heads 52 times</b>?
<p>You might be thinking to yourself, calculating that must take forever! And it would if we approached it by listing out all the possible desired outcomes with H's and T's, like we did previously, when the number of trial was only 3.
<p>Luckily, there is a faster way! How to quickly calculate binomial probabilities for large number of trials? We can apply the following formula, where...</p>


<b><font color = "darkorange">(1) - - - - - - - - - -</font></b>
$$P(X) = \ _nC_x p^x q^{n - x}$$

-  P : Probability
-  X : random variable
-  n : number of trials
-  x : number of success
-  p : probability of success
-  q : probability of failure


<p>From combinatorics, the expression nCx can be expands to:</p>

<b><font color = "darkorange">(2) - - - - - - - - - -</font></b>
$$\left( {\begin{array}{*{20}c}
   n  \\
   x  \\
\end{array}} \right) = \frac{{n!}}{{x!\left( {n - x} \right)!}}$$

<p>Then from the formula (1), we replaced (2) into (1), and generate formula (3)</p>

<b><font color = "darkorange">(3) - - - - - - - - - -</font></b>
$$P(X) = \frac{{n!}}{{x!\left( {n - x} \right)!}} p^x q^{n - x}  = \left( {\begin{array}{*{20}c}
   n  \\
   x  \\
\end{array}} \right) p^x q^{n - x}$$

### <font color = "darkcyan">2.2.1 Solution: Applying the formula of </font><font color = "darkblue">Binomial Distribution</font>

<p>Let's go back to our previous question. What is the probability that we flip a coin 100 times, and it lands heads exactly 52 out of the 100 flips? Let's first define the value of each term in our formula:</p>

-  <b>P(X): Probability of flipping 100 coins and getting heads</b>
-  since we are flipping the coin 100 times, we have 100 independent trials, therefore <b>n = 100</b>.
-  since we defining success to be "the coin get heads", and we are calculating the probability of getting 52 heads, so <b>x = 52</b>.
- because we only have heads or tails, thus the probability of success is, <b>p = 0.5</b>

<p>Plugging into our formula, we get:</p>

$$P(X) = \frac{{100!}}{{52!\left( {100 - 52} \right)!}} {(0.5)^{52} (1 - 0.5)^{100 - 52}} = 0.074$$

```python
import math

def nCx(n, x):
    f = math.factorial
    return f(n) / f(x) / f(n-x)

nCx(100, 52)*(0.5**(52))*(0.5**(48))
```

### <font color = "darkcyan">2.2.2 Solution: Use scipy to get the probability of </font><font color = "darkblue">Binomial Distribution</font>

```python
import scipy.stats as st

x = 52
n = 100
p = 0.5
st.binom.pmf(x, n, p)
```

### <font color = "darkcyan">2.2.3 Solution: Use matplotlib.pyplot to graphically present the data </font><font color = "darkblue">Binomial Distribution</font>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as st

x = 52
n = 100
p = 0.5
probability = st.binom.pmf(x, n, p)
data = st.binom.rvs(n, p, size = 100)
data.sort()
plt.bar(data, st.binom.pmf(data, n, p), linewidth = 10, color = "darkcyan")
plt.xlabel("Number of getting Heads after flipping a coin 100 times", fontsize = 12)
plt.ylabel("Probability of getting Heads", fontsize = 12)
plt.xlim(30, 70); plt.ylim(0.00, 0.08)
plt.xticks(size = 10); plt.yticks(size = 10)

plt.show()
```

## <font color = "slateblue">3.0 Mean, Variance and Standard Deviation of a </font><font color = "darkblue">Binomial Distribution</font>

### <font color = "darkcyan">3.1 Definition</font>

If X is a <b><font color = "darkblue">binomial discrete random variable</font></b> such that X ~ B(n, p), then:
-  Mean of X is equal to <b>np</b><br/>
-  Variance of X is equal to <b>npq</b><br/>
-  Standard Deviation of X is equal to <b>sqrt(npq)</b><br/>


### <font color = "darkcyan">3.2 Example 1</font>

<p>By referring to the previous question in <b>2.2 Example 1</b>, find the mean, variance and standard deviation of X.</p>

```python
import scipy.stats as st
import math

x = 52
n = 100
p = 0.5
probability = st.binom.pmf(x, n, p)
data = st.binom.rvs(n, p, size = 100)
data.sort()

mean = n * p
variance = n * p * (1 - p)
stddev = math.sqrt(variance)

print (mean)
print (variance)
print (stddev)
```
