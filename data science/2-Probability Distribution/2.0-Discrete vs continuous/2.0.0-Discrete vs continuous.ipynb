## Discrete Random Variables

A Random Variable is a set of possible values from a random experiment. 
<img src="figure1.PNG">


### Probability density function
The probability density function (PDF) of a random variable is a function describing the probabilities of each particular event occurring. <img src="figure5.PNG">

For example, a random variable describing the result of a single dice roll has the p.d.f. <img src="figure6.PNG">

In general, the value of the p.d.f. at any point must be <b>nonnegative</b> (since a negative probability is impossible), and the sum of the probabilities must be equal to 1 (as exactly one outcome must occur).


## Continuous Random Variables
<br><img src="Figure3.PNG">
### Probability density function
A continuous random variable X is given by its probability function (pdf), which is specified for the range of values for which x is valid.

Example: <img src="figure4.PNG">
Lets draw the pdf for this example.

```python
import numpy as np
import matplotlib.pyplot as plt

#step1: create a list for the values of x
x = np.arange(0, 10, 0.1)

#step2: create a list for the values of f(x)
y = 0.2-0.02*x

#step3: draw the graph
plt.plot(x, y)
```
