```python
#Data derived from https://www.weather.gov/media/sju/climo/stats/2010s.pdf
import pandas as pd
per_data = pd.read_excel('percipitation_data.xlsx', header=1, skipfooter=2)
#Let's get the data by column in a dictionary and later combine all into a single the lists
per_data.index = per_data['Months']
per_data.drop('Months' , axis='columns', inplace=True)
year_2010 = per_data[2010] #Containes the data(rows) for 2010 column\
per_data
```

```python
from fitter import Fitter
fitter_for_2010 = Fitter(year_2010)
fitter_for_2010.fit()
```

```python
fitter_for_2010.summary()
```

```python
params_of_2010 = fitter_for_2010.get_best()

param_of_expon = params_of_2010['exponpow']
```

```python
from matplotlib import pyplot as plt
import scipy.stats as st
import numpy as np

expon_power_dist = st.exponpow #Creating the distribution
x = np.linspace(st.exponpow.ppf(0.01, param_of_expon[1]), st.exponpow.ppf(0.99, param_of_expon[1]), 12)

#expon_dist_pdf = expon_power_dist.pdf(x, *param_of_expon)
plt.plot(x, st.exponpow.pdf(x, 2.7))
#plt.hist(year_2010, density=True, alpha=0.2)
plt.show()
```
