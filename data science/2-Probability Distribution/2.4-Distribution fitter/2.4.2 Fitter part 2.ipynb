## Usage
We are going to use same Database of Fitness Customers which has records such as Name, Weight, Height
<img src="fitness_table.png">

```python
#Let's import PANDAS which is another python library and it will help us to READ data from many formats, but 
#we will be reading data from only Excel file which was used in part_1 as well.
import pandas as pd  #Importing pandas

fitness_db = pd.read_excel('sample_data_bmi.xlsx', "Sheet1", skiprows=1)
#Here I am reading excel file and storing into a variable called "fitness_db"
#So "read_excel" is a function that does the job and we need to provide NAME of excel file and which sheet(2 important parameters)

weight = fitness_db['Weight'] #Storing "Weight" data into a list
height = fitness_db['Height'] #Storing "Height" data into a list
```

# Choosing the best Distribution for 'Height'

```python
fitter_for_height = Fitter(height)
#Step 1: First fitter will be for height data and we will see which distributions fit our data with the least error
fitter_for_height.fit()
```

```python
fitter_for_height.summary() #Summary shows the error and the graph of our data along with the distributions
```

As we can see from the graph provided that best fitted distribution is <b>R distribution</b> because it has the least sum_square error among all 80 distribution.

## Getting parameters of the distribution
Before we have used a function called <b>fitted_param[name_of_disttribution]</b> to get the parameters of the distribution we chose.
<br>But there is also another function to get the parameters of the best fitted distribution among all of them and it is <b>get_best()</b> function which does not require the distribution name unlike "fitted_param(name_of_disttribution)" which requires you to input name of distribution as parameter.

```python
fitter_for_height.get_best() #returns dictionary. 
#Follow the link if you are not sure what is Dictionary in python(https://www.w3schools.com/python/python_dictionaries.asp)
```

```python
#let's store the values into a list which will be used to plot the distribution
rdist_parameters_dict = fitter_for_height.get_best()
rdist_param = rdist_parameters_dict['rdist']
```

## Plotting the Probability Distribution
We will be using Matplotlib package to plot the Distribution and use Scipy package to create the distribution

```python
#Here is the plot of R Distribution for Weight dataset
from matplotlib import pyplot as plt
import scipy.stats

r_dist = scipy.stats.rdist #Creating Normal Distribution

X = linspace(153,189,100) #this is the "groundLine". Tip: you can also use min() and max() of the data here.
pdf_fitted = r_dist.pdf(X, *rdist_param)
plt.plot(X, pdf_fitted)
plt.show()
```

# Choosing the best Distribution for 'Weight'

```python
fitter_for_weight = Fitter(weight)
#Step 1: First fitter will be for height data and we will see which distributions fit our data with the least error
fitter_for_weight.fit()
```

```python
fitter_for_weight.summary() #Summary shows the error and the graph of our data along with the distributions
```

As we can see from the graph provided that best fitted distribution is <b>dgamma distribution</b> because it has the least sum_square error among all 80 distribution.

## Getting parameters of the distribution
Let's use <b>best_fit() function again</b>

```python
fitter_for_weight.get_best() #returns dictionary. 
#Follow the link if you are not sure what is Dictionary in python(https://www.w3schools.com/python/python_dictionaries.asp)
```

```python
#let's store the values into a list which will be used to plot the distribution
dgamma_parameters_dict = fitter_for_weight.get_best()
dgamma_param = dgamma_parameters_dict['dgamma']
```

## Plotting the Probability Distribution
We will be using Matplotlib package to plot the Distribution and use Scipy package to create the distribution

```python
#Here is the plot of dgamma Distribution for Weight dataset
from matplotlib import pyplot as plt
import scipy.stats

dgamma_dist = scipy.stats.dgamma #Creating Normal Distribution

X = linspace(48,82,100) #this is the "groundLine"
pdf_fitted = dgamma_dist.pdf(X, *dgamma_param)
plt.plot(X, pdf_fitted)
plt.show()
```
