## Finding the best fitted Distribution 



```python
from fitter import Fitter

resident_population_data = df = df['Resident Population']
fitter_for_rp = Fitter(resident_population_data)
fitter_for_rp.fit()
```

```python
fitter_for_rp.summary()
```

```python
fitter_for_rp.get_best()
```

```python
kappa_parameters_dict = fitter_for_rp.get_best()
kappa_param = kappa_parameters_dict['kappa4']
```

## Prediction

```python
from matplotlib import pyplot as plt
import scipy.stats as st
import numpy as np

kappa_dist = st.kappa4 #Creating the Distribution

x = np.linspace(kappa_dist.ppf(0.01, 0.1, 0), kappa_dist.ppf(0.99, 0.1, 0), 100)
X = np.linspace(308000000,330100000)
#308758105.0 starting point. 330,073,471 Ending point.



pdf_fitted = kappa_dist.pdf(X, *kappa_param)
plt.plot(X, pdf_fitted)
plt.show()
```
