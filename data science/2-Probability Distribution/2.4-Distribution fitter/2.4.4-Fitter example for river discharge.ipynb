# example 1: we have data of river discharge and wants to check witch distribution is fit with this data?

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
import pandas as pd
from fitter import Fitter
data=pd.read_csv('discharge.csv')
print (data)

```

```python
print(matplotlib.__version__)
```

```python
f = Fitter(data)

f.fit()
# may take some time since by default, all distributions are tried
```

```python
f.summary()
plt.show()
```

the result shows that the best distribution is invgauss now we want to know what is the parameters of this distribution?

```python
#Getting parameters of the distribution
f.get_best()
```

## examole 2: we have data for minimum temprature of australia and wants to check witch distribution is fit to this data?

```python

import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
import pandas as pd
from fitter import Fitter
data=pd.read_csv('daily-minimum-temperatures-Australia.csv')
print (data.temp)
tempfit = Fitter(data.temp)
tempfit.fit()

```
