```python
normdist_dict=fitter_fit.get_best()
normdist_param=normdist_dict['norm']
```

```python
norm_dist = scipy.stats.norm

x = np.linspace(norm_dist.ppf(0.01,*normdist_param), norm_dist.ppf(0.99, *normdist_param))
plt.plot(x, norm_dist.pdf(x, *normdist_param), 'g-', lw=1, alpha=1.0, label='norm pdf')

plt.show()
```
