# 1.2) Mean
1.1 Arithmetic Mean is sum of collection of numbers divided by quantity of numbers in the collection. 
    Formula: "Sum/Count" where count is quantity of numbers in the collection <br>
    <img src="meanExample.bmp">

<img src="mean.jpg">

```python
from random import randint
collectionOfNumbers = []
for i in range(100):
    num = randint(1, 100)
    collectionOfNumbers.append(num)
print(collectionOfNumbers)

"""Above I have generated 100 random numbers from 1 to 100 in the list called "collectionOfNumbers" 
To calculate the Arithmetic mean, we would DIVIDE "sum"-sum of all numbers TO "len()"-quantity of numbers in the list"""
```

```python
arithmeticMean = float(sum(collectionOfNumbers))/len(collectionOfNumbers)
print(arithmeticMean)
#So, the above result is the Arithmetic mean of that list of numbers
```
