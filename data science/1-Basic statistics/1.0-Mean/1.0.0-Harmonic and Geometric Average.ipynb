## WHAT IS <font color='springgreen'>Arithmetic mean</font>?

## <font color='cornflowerblue'>Adding all the n numbers together and then divide by the n</font>

# $$\overline{x} = \frac{\sum{x}}{n} $$

```python
# x=[1,2,4]   # store x values into a list
x=[90,1,92,85,89,94,15,88,91,8]
n= len(x)   # find the number of size and assign it to a variable n


sum=0
for i in x:
    sum+=i   # sum up the total of each x's reciprocal
print(sum/n)   # calculate the value of arithmetic mean
 
#numpy library
import numpy as np
print(np.mean(x))
```

## WHAT IS <font color='springgreen'>Harmonic mean</font>?
## <font color='cornflowerblue'>The reciprocal of the arithmetic mean of the reciprocals. Reciprocal is the multiplicative inverse of the fraction or the fraction turned upside-down.</font>


## <font color='red'>Explaination:</font>
## For example, we want to find the <font color='orange'>harmonic mean of 1, 2 and 4.</font>

## Firstly, we calculate the <font color='orange'>reciprocal of each value.</font>

## <font color='blue'>$$\frac{1}{1}=1, \frac{1}{2}=0.5, \frac{1}{4}=0.25$$</font>

## Then, <font color='orange'>add them up together.</font>

## <font color='blue'>$$ 1+0.5+0.25 = 1.75 $$ </font>

## Find the <font color='orange'>average</font> by dividing the sum with the sample size, n. In this case, it is 3.

## <font color='blue'>$$ average = \frac{1.75}{3} $$ </font>

## <font color='orange'>Reciprocal the average value</font> that we found. Done.

## <font color='blue'>$$ \frac{3}{1.75} = 1.714 $$ </font>

## <font color='RED'>THIS BRINGS TO AN EASIER FORMULA TO BE APPLIED: </font>

# $$\overline{x} = \frac{n}{\frac{1}{x_1} + \frac{1}{x_2} + \frac{1}{x_3}... + \frac{1}{x_n}} $$

## <font color='limegreen'>BUT...HOW TO DO IT IN PYTHON?</font>

## <font color='firebrick'>Step 1:</font>
<font size = 4 >Store the x values into a list. Then, find the sample size (number of terms) using len() function.</font>

```python
x=[1,2,4]  
n= len(x)  
```

## <font color='firebrick'>Step 2:</font>
<font size = 4 >Find the value of each reciprocal of x. Then, store it inside a separate list. You can use for loop to calculate each value of x's reciprocal.</font>

```python
xR=[]

for i in x:
    xR.append(1/i)
print(xR)
```

## <font color='firebrick'>Step 3:</font>
<font size = 4 >As we want to find the average, we have to find the sum of all the value of reciprocal of xs. To do this, you must first declare a variable for sum that equals to 0 and then use for loop to find the sum.</font>

```python
sum=0
for i in xR:
    sum = sum + i
```

## <font color='firebrick'>Step 4:</font>
<font size = 4 >Using formula $\frac{n}{\sum{\frac{1}{x}}}$ find the value of harmonic mean.</font>

```python
print(n/sum)
```

```python
# x=[15,30,45]   # store x values into a list
x=[90,1,92,85,89,94,15,88,91,8]
n= len(x)   # find the number of size and assign it to a variable n

xR=[]   # empty list for the reciprocal of x's (1/x)

for i in x:
    xR.append(1/i)   # calculate x's reciprocal & store into its list
print(xR)

sum=0
for i in xR:
    sum+=i   # sum up the total of each x's reciprocal
print(n/sum)   # calculate the value of harmonic mean
```

## WHAT IS <font color='springgreen'>Geometric mean</font>?

## <font color='cornflowerblue'>For n numbers, multiply the total of n numbers all together, and then take the nth root.


```python
%%HTML
<iframe width="900" height="500" src="https://www.youtube.com/embed/PKWVAIP17pw" frameborder="0" allow="accelerometer; autoplay; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>
```

## <font color='red'>Explaination:</font>
## For example, we want to find the <font color='orange'>geometric mean of 1, 2 and 4.</font>

## Firstly, we calculate the <font color='orange'> product of the x values.</font>

## <font color='blue'>$$ 1*2*4 = 8 $$</font>

## Then, find the <font color='orange'>average</font> by rooting the product of x values to the $n^{th}$. In this case, it is 3.

## <font color='blue'>$$ average = \sqrt[3]{8} $$ </font>
## <font color='blue'>$$ average = 2 $$ </font>

## <font color='RED'>SO, WE CAN CONCLUDE THAT THE FORMULA IS AS BELOW: </font>

# $$\overline{x} = \sqrt[n]{x_1*x_2*x_3...*x_n} $$

## <font color='limegreen'>LET'S DO IT USING PYTHON!</font>

## <font color='firebrick'>Step 1:</font>
<font size = 4 >Store the x values into a list. Then, find the sample size (number of terms) using len() function. </font>

```python
x=[1,2,3,4]
n= len(x)
```

## <font color='firebrick'>Step 2:</font>
<font size = 4 >Find the product of all x values. Note that you have to intialize the product of x to 1 first. Then, use for loop to calculate the product of all x values.</font>

```python
xP=1

for i in x:
    xP*=i
```

## <font color='firebrick'>Step 3:</font>
<font size = 4 >Root the product of x to the nth root, depends on the n value. You can use the formula of **power of (1/n)** for the nth root. </font>

```python
gm = xP**(1/n) 
print("Product: "+str(xP))
print("Geometric mean: "+str(gm))
```

```python
x=[1,2,4]   # store x values into a list
n= len(x)   # find the number of size and assign it to a variable n

xP=1   # initialize the product of x's as 1

for i in x:
    xP*=i   # multiply all the x's values together
gm = xP**(1/n)   # nth root the product of all x values
print("Product: "+str(xP))
print("Geometric mean: "+str(gm))
```
