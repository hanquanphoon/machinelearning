### Scatter Plot 
A scatter plot has points in the graph that shows relationship between two data using Cartesian(x,y) Coordinates.<br>
Following is the example of a Scatter Plot: <img src="ScatterPlotExample.bmp">
Now let's try to draw the Scatter Plot with Python. Refer to the Steps before start coding: <br>
1) "Import matplotlib.pyplot as plt" (this is the Python library that will help us to draw graphs)(just copy and paste it)<br> 
2) Initialize a list which contains all the ice cream sales<br>
3) Initialize a list which consists of the temperature <br>
4) Structure the ScatterPlot->plt.scatter(list1, list2, labels='skitscat')<- with the help of lists you created in step 2,3<br>
5) Finally, show the Scatter Plot with the code "plt.show()" <br>
Tip: To change color, markerType or size, use the following code while structuring the graph: color = "", marker = "", s = int (respectively)<br>
Below is the solution code you can refer to:<br>

```python
import matplotlib.pyplot as plt
temp_list = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales_list = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
plt.scatter(temp_list, sales_list, label="a")
plt.xlabel('Temperature(in Celsius)')
plt.ylabel('Sales(in $)')
plt.show()
```
