```python
import numpy as np
import matplotlib.pyplot as plt
```

# Preparing the Data

```python
age = np.random.randint(20, 80, 100) #Parameters: 1)Lowest(inclusive). 2)Highest(exclusive). 3)Size(quantity)
income = np.random.randint(5000, 50000, 100)
```

```python
#Divide "age" and "income" into groups as following
age_20_30, age_30_40, age_40_50, age_50_60, age_60_80 = [], [], [], [], []
for i in age:
    if i<30:
        age_20_30.append(i)
    elif 30<=i<40:
        age_30_40.append(i)
    elif 40<=i<50:
        age_40_50.append(i)
    elif 50<=i<60:
        age_50_60.append(i)
    else:
        age_60_80.append(i)
        
income_10k, income_20k, income_30k, income_40k, income_50k = [], [], [], [], []
for i in income:
    if i<10000:
        income_10k.append(i)
    elif 10000<=i<20000:
        income_20k.append(i)
    elif 20000<=i<30000:
        income_30k.append(i)
    elif 30000<=i<40000:
        income_40k.append(i)
    else:
        income_50k.append(i)
        
```

## Initial Step

```python
fig, ax1 = plt.subplots()
labels = '20 to 30', '30 to 40', '40 to 50', '50 to 60', '60 to 80'
sizes = [len(age_20_30), len(age_30_40), len(age_40_50), len(age_50_60), len(age_60_80)]
ax1.pie(sizes, labels=labels)
plt.show()
```

## 1. Customizing colors

```python
#Create a list of colors and pass it as a parameter
colors = ['yellow', 'green', 'blue', 'red', 'violet'] 
plt.pie(sizes, labels=labels, colors=colors)
plt.show()
```

## 2. Labelling the wedge

```python
# Pass autopct='%.1f%%' as a parameter. 0.1f is format and it shows FLOAT with one value
plt.pie(sizes, labels=labels, colors=colors, autopct='%.1f%%')
plt.show()
```

## 3. Separating the wedge

```python
#Create tuple named "explode". 0 means no change and 0.1/0.2 is mostly used separation.
explode = (0, 0.1, 0, 0, 0)
plt.pie(sizes, labels=labels, colors=colors, autopct='%.1f%%',  explode=explode)
plt.show()
```

## 4. Adding shadow

```python
#shadow=True. Pass it as a parameter
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%.1f%%', shadow=True)
plt.show()
```

## 5. Specifying the angle

```python
#startangle is another parameter which rotates the start of the pie chart based on angle provided
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%.1f%%', startangle = 90)
plt.show()
```

## 6. Making sure pie is drawn as a circle

```python
# Equal aspect ratio ensures that pie is drawn as a circle
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, startangle = 90, autopct='%.1f%%')
plt.axis('equal') ###Here
plt.show()
```

## 7. Adding title

```python
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%.1f%%', startangle = 90)
plt.axis('equal')
plt.title('Age from 20 to 80')
plt.show()
```

## 8. Making donut shape

```python
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, startangle = 90, autopct='%.1f%%')
plt.axis('equal')
plt.title('Age from 20 to 80')

#we can change pie chart to donut chart by drawing a circle with white color at origin
centre_circle = plt.Circle((0,0),0.80,fc='white') #Creating circle
fig = plt.gcf() #getting the currect figure 
fig.gca().add_artist(centre_circle) #adding the circle

plt.show()
```

## 9. Adjusting the labels

```python
#Add "pcddistance" as a parameter, default is 0.6
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, startangle = 90, autopct='%.1f%%', pctdistance=0.8)
plt.axis('equal')
plt.title('Age from 20 to 80')

#we can change pie chart to donut chart by drawing a circle with white color at origin
centre_circle = plt.Circle((0,0),0.60,fc='white') #Creating circle
fig = plt.gcf() #getting the currect figure 
fig.gca().add_artist(centre_circle)


plt.show()
```
