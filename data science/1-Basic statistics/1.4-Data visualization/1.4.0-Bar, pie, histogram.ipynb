### Bar Chart
Bar chart(or bar graph) is a graphical display of data using bars of different heights. Best way of learning is practicing and trying things out, so let's go with examples:
<br>Imagine you did a survey of your friends in school to find their favourite subject and the Data you have gathered looks like this: <img src="BarchartTable.bmp">
Let's analyze the data in the table first. So, there are four subjects(Maths, Physics, Biology and History) and (10+5+6+8) 29 students voted for their favourite subject. 10(most) students voted Maths as their favourite subject. And, we can show that in bar chart like this: <img src="BarchartGraph.bmp">
We can easily draw above bar chart with the help of Python after looking at the following second example: <img src="BarchartExample.bmp">
Now, let's try to draw the bar charts with Python. There will be steps you should follow and the (python)code is provided as well. Let's start with the first example:
<br>1) "Import matplotlib.pyplot as plt"  (this is the Python library that will help us to draw graphs)(just copy and paste it)
<br>2) Initialize a list which contains "number of votes" of each subjects
<br>3) Initialize a list which consists of all the four "subjects"
<br>4) Structure the bar chart with the help of lists you created like this "plt.bar(list1, list2)"
<br>5) Finally, show the bar chart with the code "plt.show()"
<br>Below is the solution code you can refer to:

```python
import matplotlib.pyplot as plt
```

```python
votes = [10, 5, 6, 8]#number of votes
subjects = ["Maths", "Physics", "Biology", "History"]#list of subjects
plt.bar(subjects, votes)#getting the bar chart
plt.ylabel('Number of votes') #assigning name for Y-axis (optional)
plt.title('Favourite Subject Survey') #assigning name for the bar chart (optional)
plt.show() #displaying the bar chart
```

### Histogram

Histogram is like Bar Chart, but Histogram groups numbers into ranges. Moreover, <b>Histograms</b> are great way to show results of  <b>Continues data</b> such as weight, height, length and etc. Whereas <b> Bar Charts </b> are good for <b>categirocal</b> data such as Country, Favourite Subjects and etc. Let's start with examples again:
<img src="HistogramExamplee.bmp">
Now let's try to draw the Histogram with Python. Refer to the Steps before start coding: <br>
1) "Import matplotlib.pyplot as plt" (this is the Python library that will help us to draw graphs)(just copy and paste it)<br> 
2) Initialize a list which contains all the "tree height"<br>
3) Initialize a list(called bins) which consists of the ranges such as 100, 150 and etc <br>
4) Structure the Histogram->plt.hist(list1, list2)<- with the help of lists you created in step 2,3<br>
5) Finally, show the Histogram with the code "plt.show()" <br>
Below is the solution code you can refer to:<br>


```python
treeHeight = [170, 160, 190, 180, 110, 120, 140, 240, 230, 250, 270, 260, 300, 340, 320, 280, 285]
bins = [height for height in range(351) if height>=100 and height%50==0] #or [50, 100 .. 350] 
plt.hist(treeHeight, bins)
plt.xlabel('Height(cm)')
plt.ylabel('Trees')
plt.show()
```

### Pie Chart
Pie chart is a chart that looks like "pie slices" to display relative sizes of data.<br>
Imagine you survey your friends to find the kind of movie they like the best. <img src="PychartTable1.bmp">
You can show the Data by the following Pie Chart:<img src="PiechartExample.bmp">

Now let's try to draw the Pie Chart with Python. Refer to the Steps before start coding: <br>
<br>1) "Import matplotlib.pyplot as plt" (this is the Python library that will help us to draw graphs)(just copy and paste it)</br> 
<br>2) Initialize a list which contains all the votes for each kind of movie.</br>
<br>3) Initialize a list which consists of the types of movies such as Comedy, Drama and etc </br>
<br>4) Structure the PieChart->plt.pie(list1, labels=list2)<- with the help of lists you created in step 2,3</br>
<br>5) Finally, show the pie chart with the code "plt.show()" </br>
Tip: To add percentage for each pie add the following code <b>autopct='%1.1f%%'</b> while structuring pie chart(step 4)<br>
Tip: To show data with certain colors, create a list with number of colors(same number of movie types) and include the list while structuring your pie chart with the following code <b>colors=colorList</b>(where colorList is the one you created).<br>
Below is the solution code you can refer to:<br>


```python
slices = [4, 4, 1, 6, 5]
typesOfMovies = ['Comedy', 'Scifi', 'Drama', 'Romance', 'Action']
cols = ['y', 'c','m','r','b'] #optional

plt.pie(slices, labels=typesOfMovies, colors = cols, autopct='%1.1f%%')
plt.title('Favorite Type of Movie')
plt.show()
```
