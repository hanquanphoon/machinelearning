```python
import numpy as np
import matplotlib.pyplot as plt
```

# Preparing the Data

```python
age_1 = np.random.randint(20, 80, 20) #Parameters: 1)Lowest(inclusive). 2)Highest(exclusive). 3)Size(quantity)
income_1 = np.random.randint(5000, 50000, 20)
```

```python
age_2 = np.random.randint(20, 80, 20) #Parameters: 1)Lowest(inclusive). 2)Highest(exclusive). 3)Size(quantity)
income_2 = np.random.randint(5000, 50000, 20)  
```

## Initial Step

```python
fig, ax1 = plt.subplots()
ax1.scatter(age_1, income_1, label='Scatter_1')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```

## 1. Multiple scatter plots

```python
fig, ax1 = plt.subplots()
ax1.scatter(age_1, income_1, label='Scatter_1')
ax1.scatter(age_2, income_2, label='Scatter_2') ###Just create another scatter plot
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```

## 2. Marker styling

```python
#Color. name of the color or can be first letter of the color or RGB numbers
#Marker=''. Visit the following link for markerStyles: https://matplotlib.org/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle
fig, ax1 = plt.subplots()
ax1.scatter(age_1, income_1, label='Scatter_1', color='r', marker='*')
ax1.scatter(age_2, income_2, label='Scatter_2', color='blue', marker='v')
plt.xlabel('Age')
plt.ylabel('Income')
ax1.legend()
plt.show()
```

## 3. Customizing size of marker

```python
fig, ax1 = plt.subplots()
ax1.scatter(age_1, income_1, label='Scatter_1', color='r', marker='*', s=300) #s=int, for changing marker size
ax1.scatter(age_2, income_2, label='Scatter_2', color='blue', marker='v')
plt.xlabel('Age')
plt.ylabel('Income')
ax1.legend()
plt.show()
```

## 4. Example of plotting along with GOOD programming skills 

```python
# The example was retrieved from the following link: https://pythonspot.com/matplotlib-scatterplot/

# Create data. Data can be stored in tuple or list depending on preference
N = 60
g1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))
g2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N))
g3 = (0.3*np.random.rand(N),0.3*np.random.rand(N))

data = [g1, g2, g3]
colors = ["red", "green", "blue"]
groups = ["coffee", "tea", "water"]

# Create plot
fig, ax = plt.subplots()

for data, color, group in zip(data, colors, groups):
    x, y = data
    ax.scatter(x, y, c=color, label=group)

plt.title('Matplot scatter plot')
plt.legend()
plt.show()
```

## 5. Multiple plots in a single figure

```python
#First plot
fig, ax1 = plt.subplots()
ax1.scatter(age_1, income_1, label='Scatter_1')
ax1.scatter(age_2, income_2, label='Scatter_2') ###Just create another scatter plot
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()

#Second plot
fig, ax2 = plt.subplots()
ax2.scatter(age_1, income_1, label='Scatter_1', color='r', marker='*')
ax2.scatter(age_2, income_2, label='Scatter_2', color='blue', marker='v')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()

#Third plot
# Create data
N = 60
g1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))
g2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N))
g3 = (0.3*np.random.rand(N),0.3*np.random.rand(N))

data = [g1, g2, g3]
colors = ["red", "green", "blue"]
groups = ["coffee", "tea", "water"]

# Create plot
fig, ax3 = plt.subplots()

for data, color, group in zip(data, colors, groups):
    x, y = data
    ax3.scatter(x, y, c=color, label=group)
    
plt.title('Matplot scatter plot')
plt.legend()
plt.show()
```
