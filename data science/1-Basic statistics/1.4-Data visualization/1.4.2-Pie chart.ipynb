## How to create a Pie Chart?
---
Imagine you survey your friends to find the kind of movie they like best:

![Image](table-1.PNG)

1. First, we need to find the total of the data.
   
   Add up all the numbers: **4 + 6 + 4 + 1 + 5 = 20**

2. Next, find the percentage for each of the types by 

   #### dividing the value by 20 and multiply by 100%.
![Image](table-2.PNG)

3. Now, we need to divide the circle into “pie slice” for each of the topics.

   We need to find the degrees for each sector by 

   #### dividing the value by the total, 20 and multiply by a full circle which is 360°.
   
   We will get the result below:
![Image](table-3.PNG)

Then, we can start drawing the pie chart using a protractor (manually in paper : ) ). Then color each of the sectors accordingly by using different colors and label it. You will get something like the image below:

## Drawing a pie chart using Python!
---
We have many ways to draw a pie chart by using python. Using the question above as an example.

1. One of the ways is using **python turtle**:  
![Image](turtle-pie-chart.png)
2. Another way is using **Mathplotlib**:

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Comedy', 'Action', 'Romance', 'Drama', 'SciFi'
sizes = [20, 25, 30, 5, 20]
colors = 'red','lime','yellow','gray','dodgerblue'
explode = (0.15, 0.1, 0.1, 0.8, 0)  # only "explode" the 2nd slice (i.e. 'Action')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%2.2f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
