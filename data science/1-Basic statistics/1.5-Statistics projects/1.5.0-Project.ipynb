# project 1
1- import the library that need for basic statistic.
<br>2- import data from this link(https://raw.githubusercontent.com/plotly/datasets/master/2010_alcohol_consumption_by_country.csv)</br>
<br>3- compute statistics parameters such as the mean, standard deviation, median, mode, maximum and minimum, range,1st,second and third percentile of the data</br>

# Answer:

```python
import pandas as pd
import numpy as np
from scipy import stats
data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2010_alcohol_consumption_by_country.csv')
df = data[0:10]
print(df)
```

```python
#statistic of 10 rows of data
mean = np.mean(df['alcohol'])
std = np.std(df['alcohol'])
median = np.median(df['alcohol'])
maximum = np.max(df['alcohol'])
minimum = np.min(df['alcohol'])
rg =np.ptp(df['alcohol'])
mode = stats.mode(df['alcohol'])

print("mean=", mean)
print("std=", std)
print("median=", median)
print("maximum=", maximum)
print("minimum=", minimum)
print("range=", rg)
print("mode=", mode)
```

```python
#statistic of all data
mean = np.mean(data['alcohol'])
std = np.std(data['alcohol'])
median = np.median(data['alcohol'])
maximum = np.max(data['alcohol'])
minimum = np.min(data['alcohol'])
mode = stats.mode(data['alcohol'])
rg =np.ptp(data['alcohol'])
percentile1=np.percentile(data['alcohol'],25)
percentile2=np.percentile(data['alcohol'],50)
percentile3=np.percentile(data['alcohol'],75)

print("mean=", mean)
print("std=", std)
print("median=", median)
print("maximum=", maximum)
print("minimum=", minimum)
print("mode=", mode)
print("range=", rg)
print("percentile1", percentile1)
print("percentile2", percentile2)
print("percentile3", percentile3)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
#1 
data2 = pd.read_csv('brain_size.csv')
print(data2)
```

```python
#2 separate data with';'
data2 = pd.read_csv('brain_size.csv',sep=";")
print(data2)
```

```python
#3 change '.' to nan values
data2 = pd.read_csv('brain_size.csv',sep=";", na_values=".")
print(data2)
```

```python
#4  fill the nan values with mean of that column
data2['Weight']=data2['Weight'].fillna(np.mean(data2['Weight']))
data2['Height']=data2['Height'].fillna(np.mean(data2['Height']))
print(data2)
```

```python
#5  finding statistic parameters of weight and height
print('weight_statistic:')

mean = np.mean(data2['Weight'])
std = np.std(data2['Weight'])
median = np.median(data2['Weight'])
maximum = np.max(data2['Weight'])
minimum = np.min(data2['Weight'])
mode = stats.mode(data2['Weight'])
rg =np.ptp(data2['Weight'])
percentile1=np.percentile(data2['Weight'],25)
percentile2=np.percentile(data2['Weight'],50)
percentile3=np.percentile(data2['Weight'],75)


print("mean=", mean)
print("std=", std)
print("median=", median)
print("maximum=", maximum)
print("minimum=", minimum)
print("mode=", mode)
print("range=", rg)
print("percentile=",percentile1)
print("percentile=",percentile2)
print("percentile=",percentile3)
```
