```python
import numpy as np
import scipy.stats as sp
from matplotlib import pyplot as plt
```

```python
#1
age = np.random.randint(20,80,80)
print(age)
```

```python
#2
income = np.random.randint(20000,100000,80)
print(income)
```

```python
#3
print(np.mean(age))
print(np.median(age))
print(np.percentile(age,50))
print(sp.mode(age))
```

```python
#4
group1 = [] #20-30
group2 = [] #31-40
group3 = [] #41-50
group4 = [] #51-60
group5 = [] #>60
```

```python
#5
for x in age:
    if 20<=x and x<=30:
        group1.append(x)
    elif 31<=x and x<=40:
        group2.append(x)
    elif 41<=x and x<=50:
        group3.append(x)
    elif 51<=x and x<=60:
        group4.append(x)
    else:
        group5.append(x)
print(group1)
print(group2)
print(group3)
print(group4)
print(group5)           
```

```python
#6
group1.sort()
group2.sort()
group3.sort()
group4.sort()
group5.sort()

print(group1)
print(group2)
print(group3)
print(group4)
print(group5) 
```

```python
#7
labels = 'group1', 'group2', 'group3', 'group4', 'group5'
sizes = [len(group1), len(group2), len(group3), len(group4), len(group5)]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0.1, 0.1, 0.1, 0.1, 0.2)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
#8
g1 = [] # 20k-40k 
g2 = [] # 40k-50k 
g3 = [] # 50k-60k 
g4 = [] # 60k-80k 
g5 = [] # >80k
```

```python
#9
for y in income:
    if y>=20000 and y<=40000:
        g1.append(y)
    elif y>=40001 and y<=50000:
        g2.append(y)
    elif y>=50001 and y<=60000:
        g3.append(y)
    elif y>=60001 and y<=80000:
        g4.append(y)
    else:
        g5.append(y)
print(g1)
print(g2)
print(g3)
print(g4)
print(g5)
```

```python
# 10
labels = '20k-40k', '40k-50k', '50k-60k', '60k-80k', '>80k'
sizes = [len(g1), len(g2), len(g3), len(g4), len(g5)]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'green']
explode = (0, 0, 0, 0, 0.1)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
