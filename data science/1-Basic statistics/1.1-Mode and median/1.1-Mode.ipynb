# example for finiding mode without using python library

         

1- Write a program, calculate how many times the terms is repeated in the list. Use the given method and coding. 

-Create the list of numbers (1,2,3,1,8,1,2,1,1)         
-Create a dictionaries which is b:         
<br>-Use the loop for counting the repetition of each item.hint:(a.count(i))</br>
<br>-show the values of dictionary</br>
       

```python
 a = [1,2,3,1,8,1,2,1,1]
```

```python
 b = { }
```

```python
 for i in a:
        b[i] = a.count(i)
        x=b.values()
print(b) 
print(x)
```

2-calculate the maximum values in the list. The max() function returns the item with the highest values or items with the highest value in an iterable.

<br>-add max() method at end of code </br> 


```python
m = max(x)
print(m)

```

-calculate the mode in the list. Mode is the number that occurs most often within a set of number. 

```python
for j in b:
    if b[j] == m:
        print(j)
```

 so it shows one is repeated 5 times and it would be the mode.

# repeat previous example for finiding mode with using python library 

```python
from scipy import stats
a = [1,2,3,1,8,1,2,1,1]
mode= stats.mode(a)
print(mode)
```

it shows the mode is 1 and its repeated 5 times.
