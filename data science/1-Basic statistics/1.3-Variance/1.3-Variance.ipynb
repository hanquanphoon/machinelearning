# 1.3)Average Deviation
Average Deviation is the average of absolute deviations from the central point. For deviation just think DISTANCE
<br>Average Deviation is also called as Mean Deviation. Formula: <img src="averageDeviationFormula.bmp">
<br> __Σ is Sigma__, which means to sum up.$\;\;\;\;\;\;$ __||__ (the vertical __bars__) mean __Absolute Value__, basically to ignore minus signs.
<br> **X  is** each __value__.$\;\;\;\;\;\;$ **μ is** the mean.$\;\;\;\;\;\;$ **N is** the number(quantity) of values
<br> But, there are 3 simple steps to calculate Average Deviation:
<br>1) Find the mean of all values in the list
<br>2) Subtract the mean from each values(ignore negative)
<br>3) Find the mean of those values
<br> For example, let's assume our list consists of following values:[3, 6, 6, 7, 8, 11, 15, 16]
<img src="AverageDeviationSteps.bmp">

<br>Question3.1 - Create a list which consists of any numbers
<br>Question3.2 - Find (Arithmetic)mean of all numbers in that list
<br>Question3.3 - Create another list which stores differences between the mean and all values of previous list(in q3.2)
<br>Question3.4 - Find mean of second list(from q3.3)
<br>Question3.5 - Create a function with one argument(list) and it should display the Average Deviation of the list(as arg)

```python
#question 3.5
def avgDeviation(my_list): #function that returns average Deviation of list of values
    mean = float(sum(my_list))/len(my_list) #getting the mean
    ad_list = [] #creating a list to store all distance from the mean
    for numbers in my_list:
        ad_list.append(abs(mean - numbers)) #adding distances to one list
    meanAD = float(sum(ad_list))/len(ad_list)#calculationg mean in the distance list(ad_list)
    return meanAD
x_list = [3, 6, 6, 7, 8, 11, 15, 16] #creating a new list with numbers to test(use) the function
print(avgDeviation(x_list)) #testing(using) the function
```

# 1.4) Variance
Variance is defined as average of the squared differences from the mean
Formula: $\;\;\;\;\;\;$<img src="varianceFormula.bmp"> 
As usual let's break it down to three simple steps:
<br>1)Find the mean
<br>2)Then for each number: find of the differences between the number and mean and then square them
<br>3)Find the mean of those squared differences
<br>For example: <img src="varianceExample1.bmp">
The heights (at the shoulders) are: 600mm, 470mm, 170mm, 430mm and 300mm
<br>1)Find the __mean__: 600+470+170+430+300/5 = 1970/5 = __394__
<br>2)Find difference between the number and square it. Differences: 206, 76, (-224), 36, (-94) <img src="varianceExample2.bmp">
Now find sum of squared values as shown: $$206^2 + 76^2 + (−224)^2 + 36^2 + (−94)^2 = 42436 + 5776 + 50176 + 1296 + 8836$$
<br>3)Find the mean of those squared differences: $$42436 + 5776 + 50176 + 1296 + 8836 / 5 = 108520 / 5 = 21704$$
<br>So, 21704 is the variance for our data


Question1.4.1: Generate random numbers in a list and Find  the Mean
<br>Question1.4.2: Find differences between values and the Mean and store differences in a list called "diff_list" 
<br>Question1.4.3: Update "diff_list" by Square(ing) each value 
<br>Question1.4.4: Finally, Find the mean of "diff_list" and store it.
<br>Question1.4.5: Create a function which calculates and returns the variance and the function accepts only one argument(parameter) which is List of numbers


```python
def variance(my_list):
    mean = float(sum(my_list))/len(my_list) #getting the mean
    distanceFromMean = [] #creating a list to store all (squared) differences from the mean
    for numbers in my_list:
        distanceFromMean.append(abs(mean - numbers)**2) #adding (sqaured) differences to "distanceFromMean" list
    meanVariance = float(sum(distanceFromMean))/len(distanceFromMean)
    return meanVariance
myList=[600, 470, 170, 430, 300]
print(variance(myList))
```
