use the brain size data set and make classification with random forest.

```python
import pandas as pd
import numpy as np
data= pd.read_csv('brain_size.csv', sep=";",na_values=('.'))
data
```

```python
data.describe()
```

```python
data['Weight']=data['Weight'].fillna((data['Weight'].mean()))
data['Height']=data['Height'].fillna((data['Height'].mean()))
data
```

# changing female and male to 0 and 1

```python
Gender = {'Male': 1,'Female': 0}
data.Gender = [Gender[item] for item in data.Gender] 
print(data) 
```

# Defining X and Y as label and features

```python
(X1,y1)=data.iloc[:,3:],data['Gender']
print(X1)
```

# Definig train and test

```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.33, random_state=0)
```

```python
print(X_train)
```

 Fitting random forest classification to data

```python
from sklearn.ensemble import RandomForestClassifier
rf1 = RandomForestClassifier(random_state=0)
rf1.fit(X_train, y_train)
```

Checking model for test data

```python
y1_pred = rf1.predict(X_test)
y1_pred
```

```python
y_test
```

# Evaluating the Algorithm
For classification problems the metrics used to evaluate an algorithm are accuracy, confusion matrix, precision recall, and F1 values. Execute the following script to find these values:


```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y1_pred))
print(classification_report(y_test,y1_pred))
print(accuracy_score(y_test, y1_pred))
```
